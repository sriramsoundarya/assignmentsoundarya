```

container = "patientdata"
storageAccount = "stockcontainerdaily"
```

```
client_id= dbutils.secrets.get(scope="stock_data", key="clientkey")
tenant_id= dbutils.secrets.get(scope="stock_data", key="tenantkey")
client_secret= dbutils.secrets.get(scope="stock_data", key="clientsecretkey")
```

```
configs = {"fs.azure.account.auth.type": "OAuth",
           "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
           "fs.azure.account.oauth2.client.id": f"{client_id}",
           "fs.azure.account.oauth2.client.secret": f"{client_secret}",
           "fs.azure.account.oauth2.client.endpoint": f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"}
```

```
dbutils.fs.mount(
    source = f"abfss://{container}@{storageAccount}.dfs.core.windows.net/",
    mount_point = f"/mnt/{container}/{storageAccount}",
    extra_configs = configs
)
```

```
df=spark.read.format("csv").option("header", "true").option("delimiter","|").load(f"/mnt/{container}/{storageAccount}/raw_data/*.csv")
```

```
display(df)
```

```
df=df.withColumnRenamed("Customer_Name","Name").withColumnRenamed("Customer_ID","Cust_I").withColumnRenamed("Open_Date","Open_Dt").withColumnRenamed("Last_Consulted_Date","Consul_Dt").withColumnRenamed("Vaccination_Id","VAC_ID").withColumnRenamed("Is_Active","Flag").drop("Record_Type")
```

```
df.write.format("delta").mode("append").save(f"/mnt/{container}/{storageAccount}/staging_data")
```

```
display(df)
```

Assuming flag as Y as latest record

```
df.filter(f"Flag = 'Y'").write.format("parquet").mode("append").partitionBy("Country").save(f"/mnt/{container}/{storageAccount}/partitioned_data")
```

```
from pyspark.sql.functions import input_file_name

partitoned_df=spark.read.format("parquet").load(f"/mnt/{container}/{storageAccount}/partitioned_data")
```

```
partition_values = partitoned_df.select("Country").distinct().collect()
for partition in partition_values:
    Country = partition["Country"]
    table_name = f"Table_{Country}"
    # filtered_df = partitoned_df.filter(f"Country = '{Country}'")
    partitoned_df.filter(f"Country = '{Country}'").write.format("delta").mode("append").option(
        "path", f"/mnt/{container}/{storageAccount}/tables_data/"
    ).saveAsTable(table_name)
```
